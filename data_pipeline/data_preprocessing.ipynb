```python
import pandas as pd
```

```python
df = pd.read_csv("train_data.csv")
df
```

Data points with time_to_tca within 2 days are removed from the dataset due to the constraint of the challenge to use only CDMs atleast 2 days prior to tca:

```python
challenge_df = df[df['time_to_tca'] >= 2].reset_index(drop=True)
challenge_df
```

Analysing missing values in the dataset:

```python
null_percentage = challenge_df.isnull().mean()*100
print('total rows with values in all variables: ', challenge_df.dropna().shape[0])
print('total rows in the data: ', challenge_df.shape[0])
print('percentage of data with missing values: ', (1 - challenge_df.dropna().shape[0]/ challenge_df.shape[0]))
print('Top 10 Highest percentage of missing values and corresponding columns:')
null_percentage.sort_values(ascending=False)[:10]
```

The key steps used to remove missing values are:

1. Grouping the data by the mission_id column

2. Using ffill() and bfill() to fill the missing values with the nearest non-missing value in the same mission_id

```python
#Replace missing values with nearest neighbour

imputed_df = (
    challenge_df.groupby('mission_id')
    .apply(lambda x: x.ffill().bfill())
    .reset_index(drop=True)
)
print(challenge_df.isna().sum().sum())
print(imputed_df.isna().sum().sum())
```

Handling Outliers:

```python
def detect_outliers(imputed_df):
    outlier_counts = {}
    for col in imputed_df.select_dtypes(include='number'):
        q1 = imputed_df[col].quantile(0.25)
        q3 = imputed_df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outliers = imputed_df[(imputed_df[col] < lower_bound) | (imputed_df[col] > upper_bound)].shape[0]
        outlier_counts[col] = outliers

    outlier_counts = dict(sorted(outlier_counts.items(), key=lambda x: x[1], reverse=True))
    return outlier_counts

print("Mean skewness of dataset before outlier removal:", imputed_df.skew(numeric_only=True).mean())
detect_outliers(imputed_df)
```

```python
clean_df = imputed_df.copy()
for col in imputed_df.select_dtypes(include='number'):
        q1 = imputed_df[col].quantile(0.25)
        q3 = imputed_df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        
        # Replace outliers with the lower and upper bounds
        clean_df.loc[imputed_df[col] < lower_bound, col] = lower_bound
        clean_df.loc[imputed_df[col] > upper_bound, col] = upper_bound

print("Mean skewness of outlier removed dataset: ",clean_df.skew(numeric_only=True).mean())
detect_outliers(clean_df)
```

The c_object_type is a categorical variable. The category value is replaced with the median value of risk for that particular category. 

This provides a numeric relation between the categorical variable and the target variable (i.e. risk)

```python
#category c_object_type
print(clean_df['c_object_type'].value_counts())
risk_medians = clean_df.groupby(['c_object_type'])['risk'].median()
clean_df['c_object_type'] = clean_df['c_object_type'].map(risk_medians)
```

```python
clean_df.to_csv('clean_data.csv', index=False)
```
